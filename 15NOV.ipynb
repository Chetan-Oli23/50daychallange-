```python
# Data
import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4,5],"B":[9,8,7,6,5]})
print(var)
```

```python
# Insert
var.insert(1,"C",var["A"])  # 1 is the index position where "C" is insert
print(var)
```

```python
# insert "C1" in 1 index position
var.insert(1,"C1",[11,12,13,14,15])
print(var)
```

```python
#insert 3 digits from "A"
var["C3"] = var["A"][:3]
print(var)
```

```python
# Data
var1 = pd.DataFrame({"A":[1,2,3,4,5],"B":[9,8,7,6,5],"C":[11,12,13,14,15]})
print(var1)
```

```python
#pop
var1.pop("B")
print(var1)
```

```python
# Delete
del var1["A"]
print(var1)
```
