Simple Display Data Using pandas libaray

```python
# Series
import pandas as pd
x = [3,4,5,6]
var=pd.Series(x,index=['a','b','c','d'],dtype="float",name="Python")
print(var)
print(type(var))
```

```python
dic = {"name": ['python','c','c++',"java"],"popular":[12,13,14,15],"rank":[1,4,3,2]}
var1 = pd.Series(dic)
print(var1)
```

```python
# Display 12 in 1 to 6 index
s= pd.Series(12,index=[1,2,3,4,5,6])
print(s)
print(type(s))
```

```python
# Add to Series
s1 = pd.Series(12,index=[1,2,3,4,5,6])
s2 = pd.Series(10,index=[1,2,3,])
print(s1+s2)
```

4,5,6 are NaN because series (s2) are no 4,5,6 index so that they are so NaN when add to Sereis
