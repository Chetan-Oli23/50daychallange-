```python
# 1. write program to vehicle tracker with graph

import matplotlib.pyplot as plt

# --------- Vehicle Class ----------
class Vehicle:
    def __init__(self, name):
        self.name = name        # Vehicle name
        self.speeds = []        # List to store speed readings

    # Add a speed reading
    def add_speed(self, speed):
        self.speeds.append(speed)

    # Calculate statistics
    def calculate_stats(self):
        if len(self.speeds) == 0:
            return 0, 0, 0
        max_speed = max(self.speeds)
        min_speed = min(self.speeds)
        avg_speed = sum(self.speeds) / len(self.speeds)
        return max_speed, min_speed, avg_speed

    # Display statistics
    def display_stats(self):
        max_s, min_s, avg_s = self.calculate_stats()
        print(f"\nVehicle: {self.name}")
        print(f"Maximum speed: {max_s}")
        print(f"Minimum speed: {min_s}")
        print(f"Average speed: {avg_s:.2f}")

    # Draw graph of speeds over time
    def draw_graph(self):
        if len(self.speeds) == 0:
            print("No speed data to plot.")
            return
        plt.figure(figsize=(5,3))
        plt.plot(range(1, len(self.speeds)+1), self.speeds, marker='o', linestyle='-', color='blue')
        plt.title(f"Speed of {self.name} Over Time")
        plt.xlabel("Reading Number")
        plt.ylabel("Speed")
        plt.grid(True)
        plt.show()

# --------- Main Program ----------
print("=== Vehicle Speed Tracker ===")
vehicle_name = input("Enter vehicle name: ")
vehicle = Vehicle(vehicle_name)

# Input 10 speed readings
for i in range(1, 11):
    while True:
        try:
            speed = float(input(f"Enter speed reading {i}: "))
            if speed < 0:
                print("Speed cannot be negative. Try again.")
                continue
            break
        except ValueError:
            print("Invalid input! Please enter a number.")
    vehicle.add_speed(speed)

# Display statistics
vehicle.display_stats()

# Show speed graph
vehicle.draw_graph()
```
