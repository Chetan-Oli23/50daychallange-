**Arithmetic Operators:**

1.  Addition Operations -> +
2.  Subtraction Operations -> -
1.  Multiplication Operations -> *
2.  Division Operations -> /
1.  Modulo Operations -> %
1.  Floor Division Operations -> //
2.  Power Operation -> **









```python

import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
print(var)
```

```python

# 1. Add Two List
var ["C"] = var["A"] + var["B"]
print(var)

```

```python

# 2. Subtraction Operator
var["C"] = var["B"] - var["A"]
print(var)
```

```python

# 3. Multiplication Operator
var["C"] = var["B"] * var["A"]
print(var)
```

```python
# 4. Division Operator
var["C"] = var["B"] / var["A"]

# Display
display(var.style.format({"C": "{:.2f}"}))
```

```python

# 5. Modulo Operator
var["C"] = var["B"] % var["A"]
print(var)
```

```python
# 6. Floor Division Operator
var["C"] = var["B"] // var["A"]
print(var)
```

```python

# 7. Power Operator
var["C"] = var["B"] ** var["A"]
print(var)
```
