```python
# To sort row into descending order
csv1.sort_index(axis=0,ascending=False)  # If False arrange descending order.
                                         # If True arrange ascending order.
```

```python
# To sort columns into descending order
csv1.sort_index(axis=1,ascending=False) # If False arrange descending order.
                                        # If True arrange ascending order.
```

**loc[]**
*   Select rows by label.
*   Eg: df.loc[0:5]

**iloc[]**
*   Select rows by index number.
*   Eg: df.iloc[0:3]





```python
csv1.loc[0:3]  # 0,1,2,3 rows only show
```

```python
csv1.iloc[0:3]
```
