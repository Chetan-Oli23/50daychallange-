***DataFame :*** DataFrame is widely used data structures in Pandas. DataFrame as a table or Spreadsheet in Python.

```python
# DataFrame

import pandas as pd
l = [1,2,3,4,5,6]
var=pd.DataFrame(l)
print(var)
```

```python

d = {"a":[1,2,3,4,5],"s":[1,2,3,4,5]}
var1=pd.DataFrame(d)
print(var1)
print(type(var1))
```

```python
# Columns
d={"a":[1,2,3,4,5],"s":[1,2,3,4,5], "d":[1,2,3,4,5],1:[1,2,3,4,5]}
var1=pd.DataFrame(d)
print(var1)
print(type(var1))
print(var1.columns)
```

```python
# only need columns extract
d={"a":[1,2,3,4,5],"s":[1,2,3,4,5], "d":[1,2,3,4,5],1:[1,2,3,4,5]}
var1=pd.DataFrame(d,columns=["a",1,"s"])
print(var1)
print(type(var1))
print(var1.columns)
```

```python
# Change Index Name
d={"a":[1,2,3,4,5],"s":[1,2,3,4,5], "d":[1,2,3,4,5],1:[1,2,3,4,5]}
var1=pd.DataFrame(d,columns=["a",1],index=["c","d","e",1,3])
print(var1)
print(var1.columns)
print(var1.index)
```

```python
#get data
d={"a":[1,2,13,4,5],"s":[1,2,3,4,5],"d":[1,2,3,4,5],1:[1,2,3,4,5]}
var1=pd.DataFrame(d)
print(var1)
print("Get Data From array 'a' Index 2") # Index Always  start from 0
print(var1["a"][2])
```
